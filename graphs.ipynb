```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall=[17,10,10,10,26,50]
months=['jan','feb','mar','apr','may','jun']
```

```python
plt.style.available
plt.style.use('ggplot') #theme
plt.bar(months,rainfall,color='#585555')
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
)
plt.show()
```

```python
#donut
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
    wedgeprops={'width':0.4},
    pctdistance=0.7
)
plt.show()
```
